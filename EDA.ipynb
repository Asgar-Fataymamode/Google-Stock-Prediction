```python
# GOOGLE STOCK (2004 - 2020) EDA NOTEBOOK

# 1. IMPORT LIBRARIES
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Display settings
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', 100)

# if in Jupyter or IPython
%matplotlib inline

# 2. LOAD THE DATA
file_path = 'GOOGL.csv.xls'
data = pd.read_csv(file_path)

# 3. INITIAL INSPECTION
print("         INITIAL INSPECTION      ")

print(">>> First 5 rows:")
display(data.head())

print("\n>>> Last 5 rows:")
display(data.tail())

print("\n>>> Shape of dataset:")
print(f"Rows: {data.shape[0]}, Columns: {data.shape[1]}")

print("\n>>> Dataset Info:")
print(data.info())

print("\n>>> Descriptive Statistics (numeric columns):")
display(data.describe())

print("\n>>> Column Names:")
print(data.columns.to_list())

# 4. DATA CLEANING & PREPROCESSING
print("   DATA CLEANING & PREPROCESSING  ")

# 4.1 CHECK FOR DUPLICATES
duplicates_count = data.duplicated().sum()
print(f"Number of duplicate rows: {duplicates_count}")
if duplicates_count > 0:
    print("Removing duplicates...")
    data.drop_duplicates(inplace=True)
    print("Duplicates removed.")
else:
    print("No duplicate rows found.")

# 4.2 CHECK FOR MISSING VALUES
missing_values = data.isnull().sum()
print("\nMissing values for each column:")
print(missing_values)

# 4.3 HANDLE DATA TYPES
# Convert the 'Date' column to datetime
if data['Date'].dtype == object:
    data['Date'] = pd.to_datetime(data['Date'], errors='coerce')

# Sort by date in ascending order
data.sort_values(by='Date', inplace=True)
data.reset_index(drop=True, inplace=True)

print("\n>>> Checking after date conversion and sort:")
print("Data 'Date' type:", data['Date'].dtype)
print("Dataset shape:", data.shape)

print("\n>>> Display sample data after cleanup:")
display(data.head())

# 5. UNIVARIATE ANALYSIS
print("         UNIVARIATE ANALYSIS      ")

# Identify numeric columns for convenience
numeric_cols = ['Open', 'High', 'Low', 'Close', 'Adj Close', 'Volume']

# 5.1 HISTOGRAMS & KERNEL DENSITY ESTIMATES
fig, axes = plt.subplots(nrows=len(numeric_cols), figsize=(8, 25))
for i, col in enumerate(numeric_cols):
    sns.histplot(data[col], kde=True, color='royalblue', ax=axes[i])
    axes[i].set_title(f"Distribution of {col}", fontsize=12)
    axes[i].set_xlabel(col)
plt.tight_layout()
plt.show()

# 5.2 BOX PLOTS (CHECKING FOR OUTLIERS)
fig, axes = plt.subplots(nrows=len(numeric_cols), figsize=(8, 25))
for i, col in enumerate(numeric_cols):
    sns.boxplot(x=data[col], color='lightblue', ax=axes[i])
    axes[i].set_title(f"Box Plot of {col}", fontsize=12)
    axes[i].set_xlabel(col)
plt.tight_layout()
plt.show()

# 6. BIVARIATE ANALYSIS
print("         BIVARIATE ANALYSIS       ")

# 6.1 CORRELATION MATRIX
corr_matrix = data[numeric_cols].corr()

print("\n>>> Correlation Matrix:")
display(corr_matrix)

# 6.2 CORRELATION HEATMAP
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Heatmap of Numerical Features")
plt.show()

# 7. TIME SERIES ANALYSIS
print("       TIME SERIES ANALYSIS       ")

# 7.1 CHECKING STOCK PRICE & VOLUME OVER TIME

# Plot 1: Open & Close over time
plt.figure(figsize=(14, 6))
plt.plot(data['Date'], data['Open'], label='Open', color='blue')
plt.plot(data['Date'], data['Close'], label='Close', color='red')
plt.title("Google Stock - Open vs. Close Price Over Time")
plt.xlabel("Date")
plt.ylabel("Price (USD)")
plt.legend()
plt.show()

# Plot 2: High & Low over time
plt.figure(figsize=(14, 6))
plt.plot(data['Date'], data['High'], label='High', color='green')
plt.plot(data['Date'], data['Low'], label='Low', color='orange')
plt.title("Google Stock - High vs. Low Price Over Time")
plt.xlabel("Date")
plt.ylabel("Price (USD)")
plt.legend()
plt.show()

# Plot 3: Volume over time
plt.figure(figsize=(14, 6))
plt.plot(data['Date'], data['Volume'], label='Volume', color='purple')
plt.title("Google Stock - Volume Traded Over Time")
plt.xlabel("Date")
plt.ylabel("Volume")
plt.legend()
plt.show()

# 7.2 ROLLING AVERAGES
data['Close_30d_MA'] = data['Close'].rolling(window=30).mean()

plt.figure(figsize=(14, 6))
plt.plot(data['Date'], data['Close'], label='Close Price', alpha=0.5)
plt.plot(data['Date'], data['Close_30d_MA'], label='30-day MA of Close', color='red')
plt.title("Google Stock - 30-day Rolling Average (Close Price)")
plt.xlabel("Date")
plt.ylabel("Price (USD)")
plt.legend()
plt.show()



```
