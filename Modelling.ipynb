```python
#  GOOGLE STOCK MODELLING NOTEBOOK

# 1. IMPORT LIBRARIES
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split, GridSearchCV, cross_val_score
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier

from sklearn.metrics import (accuracy_score, 
                             confusion_matrix, 
                             classification_report,
                             roc_curve, 
                             auc)

# If in a Jupyter/Colab environment
%matplotlib inline

# 2. LOAD & PREPARE THE DATA
file_path = 'GOOGL.csv.xls' 
df = pd.read_csv(file_path)

print("         INITIAL DATA INSPECTION        ")
print(">>> Head of the dataset:")
display(df.head())

print("\n>>> Info:")
print(df.info())

print("\n>>> Shape:")
print(df.shape)

# Convert 'Date' to datetime & sort
df['Date'] = pd.to_datetime(df['Date'], errors='coerce')
df.sort_values(by='Date', inplace=True)
df.reset_index(drop=True, inplace=True)

# 3. FEATURE ENGINEERING
# Create a binary target indicating if next day's close is higher than today's
df['Target'] = np.where(df['Close'].shift(-1) > df['Close'], 1, 0)
# drop last row that becomes NaN after shift
df.dropna(inplace=True)  

# Add rolling features
df['Rolling_5d_Close'] = df['Close'].rolling(window=5).mean()
df['Rolling_5d_Volume'] = df['Volume'].rolling(window=5).mean()
df.fillna(method='bfill', inplace=True)

# 4. DEFINE FEATURES (X) AND TARGET (y)
features = ['Open', 'High', 'Low', 'Close', 'Volume',
            'Rolling_5d_Close', 'Rolling_5d_Volume']
X = df[features].values
y = df['Target'].values

print("\n>>> Sample of feature-target columns:")
display(df[['Date'] + features + ['Target']].head(10))

print("\n>>> Features and Target shapes:")
print("X shape:", X.shape)
print("y shape:", y.shape)

# 5. TRAIN-TEST SPLIT (TIME-BASED)
train_size = int(0.8 * len(df))
X_train, X_test = X[:train_size], X[train_size:]
y_train, y_test = y[:train_size], y[train_size:]

print("\n>>> Train/Test Split (Time-based)")
print(f"Train set: {X_train.shape}, Test set: {X_test.shape}")

# 6. DATA SCALING
scaler = StandardScaler()
scaler.fit(X_train)

X_train_scaled = scaler.transform(X_train)
X_test_scaled  = scaler.transform(X_test)

# ADDITIONAL VISUALIZATIONS (7+ PLOTS)

# PLOT #1: DISTRIBUTION OF THE TARGET VARIABLE
plt.figure(figsize=(6,4))
sns.countplot(x=df['Target'], palette='Set2')
plt.title("Distribution of Target (Up=1 / Down=0)")
plt.xlabel("Target")
plt.ylabel("Count")
plt.show()

# PLOT #2: CORRELATION HEATMAP (FEATURES + TARGET)
corr_cols = features + ['Target']
corr_matrix = df[corr_cols].corr()

plt.figure(figsize=(8,6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Heatmap (Features + Target)")
plt.show()

# PLOT #3: FEATURE DISTRIBUTION BEFORE / AFTER SCALING (OPTIONAL)
feat_to_demo = ['Close', 'Volume'] 
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(10,8))
for i, feat in enumerate(feat_to_demo):
    # Original
    sns.histplot(X_train[:, features.index(feat)], kde=True, ax=axes[i][0], color='gray')
    axes[i][0].set_title(f"Original Distribution: {feat}")
    
    # Scaled
    sns.histplot(X_train_scaled[:, features.index(feat)], kde=True, ax=axes[i][1], color='blue')
    axes[i][1].set_title(f"Scaled Distribution: {feat}")

plt.tight_layout()
plt.show()

# 7. MODELING: LOGISTIC REGRESSION, RANDOM FOREST, XGBOOST
print("\n===========================================")
print("       BASELINE MODEL TRAINING & EVAL      ")
print("===========================================\n")

# 7.1 LOGISTIC REGRESSION
lr_model = LogisticRegression(random_state=42, max_iter=1000)
lr_model.fit(X_train_scaled, y_train)
lr_pred = lr_model.predict(X_test_scaled)

lr_accuracy = accuracy_score(y_test, lr_pred)
print(f"Logistic Regression Accuracy: {lr_accuracy:.4f}")

# 7.2 RANDOM FOREST
rf_model = RandomForestClassifier(random_state=42, n_estimators=100)
rf_model.fit(X_train_scaled, y_train)
rf_pred = rf_model.predict(X_test_scaled)

rf_accuracy = accuracy_score(y_test, rf_pred)
print(f"Random Forest Accuracy: {rf_accuracy:.4f}")

# 7.3 XGBOOST
xgb_model = XGBClassifier(random_state=42, use_label_encoder=False, eval_metric='logloss')
xgb_model.fit(X_train_scaled, y_train)
xgb_pred = xgb_model.predict(X_test_scaled)

xgb_accuracy = accuracy_score(y_test, xgb_pred)
print(f"XGBoost Accuracy: {xgb_accuracy:.4f}")

# 8. EVALUATION (with more PLOTS)
def evaluate_model(name, y_true, y_pred):
    print(f"\n=== {name} ===")
    print("Accuracy:", accuracy_score(y_true, y_pred))
    print("Confusion Matrix:\n", confusion_matrix(y_true, y_pred))
    print("Classification Report:\n", classification_report(y_true, y_pred))

evaluate_model("Logistic Regression", y_test, lr_pred)
evaluate_model("Random Forest", y_test, rf_pred)
evaluate_model("XGBoost", y_test, xgb_pred)


# PLOT #4, #5, #6: CONFUSION MATRIX PLOTS FOR EACH MODEL
from mlxtend.plotting import plot_confusion_matrix

models_preds = [("Logistic Regression", lr_pred),
                ("Random Forest",       rf_pred),
                ("XGBoost",             xgb_pred)]

for model_name, pred in models_preds:
    cm = confusion_matrix(y_test, pred)
    plt.figure(figsize=(5,4))
    plot_confusion_matrix(cm, figsize=(5,4), cmap=plt.cm.Blues, colorbar=True)
    plt.title(f"Confusion Matrix - {model_name}")
    plt.show()

# PLOT #7: ROC CURVE COMPARISON
plt.figure(figsize=(8,6))

model_dict = {
    "Logistic Regression": lr_model,
    "Random Forest": rf_model,
    "XGBoost": xgb_model
}

for name, model in model_dict.items():
    # Probability for the positive class
    y_score = model.predict_proba(X_test_scaled)[:,1]
    fpr, tpr, _ = roc_curve(y_test, y_score)
    roc_auc = auc(fpr, tpr)
    plt.plot(fpr, tpr, label=f"{name} (AUC={roc_auc:.3f})")

plt.plot([0,1],[0,1],'r--')
plt.title("ROC Curve Comparison")
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.legend(loc='lower right')
plt.show()

# PLOT #8: FEATURE IMPORTANCE (RANDOM FOREST)

# 8.1 RANDOM FOREST FEATURE IMPORTANCES
rf_importances = rf_model.feature_importances_
feat_imp_df = pd.DataFrame({
    'Feature': features,
    'Importance': rf_importances
}).sort_values(by='Importance', ascending=False)

plt.figure(figsize=(6,4))
sns.barplot(x='Importance', y='Feature', data=feat_imp_df, palette='viridis')
plt.title("Random Forest Feature Importances")
plt.tight_layout()
plt.show()

# PLOT #9: FEATURE IMPORTANCE (XGBOOST)
xgb_importances = xgb_model.feature_importances_
xgb_imp_df = pd.DataFrame({
    'Feature': features,
    'Importance': xgb_importances
}).sort_values(by='Importance', ascending=False)

plt.figure(figsize=(6,4))
sns.barplot(x='Importance', y='Feature', data=xgb_imp_df, palette='magma')
plt.title("XGBoost Feature Importances")
plt.tight_layout()
plt.show()
```
